```python
# 設置網址收集陣列
urls=[]

# 清空放置列表資訊的變數，避免重新執行這個cell時重複匯入至urls
urls.clear()

# 取得元素
div = soup.select('li.shopee-search-item-result__item a.contents[href]')
for i in div:

  print(i.get_text())
  print(i['href']) 
  print("-----------")

  # 加入列表資訊
  urls.append({
      'title': i.get_text(),
      'link': 'https://shopee.tw' + i['href']
  })

  # 生成1到10秒之間的隨機數
  random_sleep_time = randint(1, 7)

  # 讓程式隨機睡眠
  sleep(random_sleep_time)
  
```
